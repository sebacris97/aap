Investigue/repase  que son las listas, tuplas, conjuntos y diccionarios nativos de Python. Utilizando los constructores para cada tipo de dato genere códigos de ejemplo y recórralos imprimiendo sus valores.



````python
# ******************** LISTAS *********************
# Constructores de listas: [] y list()
lista1 = [1, 2, 3, 4, 5]
lista2 = list()  # lista vacia. Puede recibir un "Iterable"
lista2.append("Perro")
lista2.append("Gato")

lista1[4] = 6  # modifica el 5to elemento

print("Contenido: ", lista1, lista2, "\n")

# Recorrido de listas
for e in lista1:
    print(e, " | ", end=" ")
print("\n") # deja línea por legibilidad

for i in range(len(lista2)):
    print(i, "=>", lista2[i], " | ", end=" ")
print("\n") # deja línea por legibilidad

for i, e in enumerate(lista2):
    print(i, "=>", e, " | ", end=" ")
print("\n") # deja línea por legibilidad
````

````python
# ******************** TUPLAS *********************
# Las tuplas son inmutables. Una vez generadas no pueden modificarse!
# Constructores de tuplas: () y tuple()
tupla1 = tuple(lista2) # tupla a partir de una lista
tupla2 = ("Caballo", 3.14159, 10)

print("Contenido de tuplas: ", tupla1, tupla2, "\n")

# Error, no puede modificarse una tupla
# tuple1[1] = 5

# Error, las tuplas no son iterables
# for e in tuple:
#     print(e, ' | ', end=' ')

for i in range(len(tupla1)):
    print(i, "=>", tupla1[i], " | ", end=" ")
print("\n") # deja línea por legibilidad

a, b, c = tupla2
print(a, b, c, "\n")
````

````python
# ****************** CONJUNTOS  *********************
set1 = set([1, 1, 2, 2, 3, 4]) # conjunto a partir de una lista
set2 = {"Casa", "Auto", "Moto"}

print("Contenido de conjuntos: ", set1, set2, "\n")

# Error, no puede modificarse una tupla
# tuple1[1] = 5

# Error, las tuplas no son iterables
# for e in tuple:
#     print(e, ' | ', end=' ')

for e in set1:
    print(e, " | ", end=" ")
print("\n") # deja línea por legibilidad

for i, e in enumerate(set2):
    print(i, "=>", e, " | ", end=" ")
print("\n") # deja línea por legibilidad

set2.remove("Moto")  # elimina Moto del conjunto, si no existe da error
set2.discard("Moto") # elimina Moto del conjunto, si no existe NO da error
print(set2)
````

````python
# ****************** DICCIONARIOS  *********************
dict1 = dict()
dict1[0] = "Enero" # agrega directamente un clave => valor
dict2 = {"Casa": 10, "Auto": 20, "Moto": 30}
dict3 = dict([('a', 2), ('b', 4)]) # construye a partir de lista de tuplas

print("Contenido de diccionarios: ", dict1, dict2, dict3,"\n")

for e in dict3:
    print(e, " | ", end=" ")
print("\n") # deja línea por legibilidad

for i, e in enumerate(dict2):
    print(i, "=>", e, " | ", end=" ")
print("\n") # deja línea por legibilidad

print(dict2.keys(), dict2.values(), "\n")

print("Auto" in dict2, "\n")  # verificar si existe una clave

# Acceso a traés de claves que existen/no existen
print(dict3["a"], dict3.get("a"), dict3.get("c"), dict3.get("c", 100), "\n")

del dict2["Moto"]  # elimina moto del diccionario
print(dict2)
````

<font color=#3d85c6> <h4> <b><i>Ejercicio 3</i></b> </font>

Implemente una función que a partir de la lista que recibe cómo parámetro, retorne una nueva lista sin elementos repetidos. Compruebe su correcto funcionamiento.



````python
def elimina_rep(lst):
    lst_nue =[]
    for e in lst:
        if e not in lst_nue:
          lst_nue.append(e)
    return lst_nue

# lista con repeticiones
lst_a = [1, 2, 3, 4, 5, 6, 1, 2, 3, 4, 5, 6]
print("original:", lst_a, "\n", "  nueva:", elimina_rep(lst_a))
````

Investigue y escriba código que demuestre el funcionamiento de los “slices” en listas.


````python
# ************* SLICES sobre listas **************
# genera una lista por comprensión de los 10 primeros numeros
lst_a = [x for x in range(10)] 

# imprime elementos en un rango:
# todos los elementos
print("todos:", lst_a, lst_a[::], "\n")

# imprime 1er y último elemento
print("primero:", lst_a[0], "     último:", lst_a[-1], "\n")

# imprime 2do y anteúltimo elemento
print("segundo:", lst_a[1], " anteúltimo:", lst_a[-2], "\n")

# todos menos el primero
print("todos - primero:", lst_a[1:], "\n")

# todos menos el ultimo
print("todos - último:", lst_a[:-1], "\n")

# todos menos primero y ultimo
print("todos - primero - último:", lst_a[1:-1], "\n")

# imprime impares (incrementa de a 2)
print("impares:", lst_a[1::2], "\n")

# imprime pares (incrementa de a 2)
print("  pares:", lst_a[::2], "\n")

# imprime múltiplos de 3 (incrementa de a 3)
print("multiplos de 3:", lst_a[3::3], "\n")

# imprime todos en sentido inverso
print("inverso:", lst_a[::-1], "\n")
````

````python
import numpy as np

# ¿Qué diferencias hay entre los constructores, array, empty, full, zeros, ones, identity?

# constructor array a partir de una lista
arr_a = np.array([ [1, 2, 3], [4, 5, 6]])
print("Valores de lista:\n", arr_a, "\n")

# constructor empty. Crea un array de la dimensión indicada sin valores asignados
arr_a = np.empty((2,3)) # puede tener cualquier valor
print("Valores con basura:\n",arr_a, "\n")

# constructor full. Crea un array de la dimensión indicada con el valor indicado
arr_a = np.full((2,3), 10) 
print("Valores con 10:\n", arr_a, "\n")

# constructor zeros. Crea un array de la dimensión indicada con el valor 0
arr_a = np.zeros((2,3))
print("Valores con ceros:\n", arr_a, "\n")

# constructor ones. Crea un array de la dimensión indicada con el valor 1
arr_a = np.ones((2,3))
print("Valores con unos:\n",arr_a, "\n")

# constructor ones. Crea un array identidad de la dimensión indicada
arr_a = np.identity(3)
print("Valores con identidad:\n", arr_a, "\n")

````

Investigue el funcionamiento del Dataframe de Pandas y cree uno con la información de la siguiente tabla:


| Nombre | Edad | Pais |
| --- | --- | --- |
| Juan | 20 | Argentina |
| María | 26 | Peru |
| Pedro | 18 | Brasil |
| José | 22 | Chile |


Realice las siguientes operaciones:
* Imprimir los nombres de las columnas.
* gregar a la tabla a Pablo que tiene 30 años y es originario de Peru. Agregarlo de 2 formas diferentes.
* Eliminar de la tabla al Pedro repetido.
* Modificar los atributos de paises que dicen “Peru” (sin acento) y reemplazarlos por “Perú” (con acento).

````python
import pandas as pd

# Creación del DataFrame
# Los datos tienen que estar en un diccionario donde cada entrada
# es el título de la columna y el valor una lista con los valores 
# de dicha columna
#columnas = ["Nombre", "Edad", "Pais"]
datos = {
    "Nombre": ["Juan", "María", "Pedro", "José"],
    "Edad": [20, 26, 18, 22],
    "Pais": [ "Argentina", "Peru", "Brasil", "Chile"]
    }

df = pd.DataFrame(data=datos)

print(df, "\n")

# Agregar una fila
persona = {"Nombre": "Pablo", "Edad": 30, "Pais": "Peru"}
nuevo_df = df.append(persona, ignore_index=True) # es obligatorio el índice con V
nuevo_df.loc[nuevo_df.shape[0]] = ["Pablo", 30, "Peru"]
# El dataframe no se modifica. Crea una nueva instancia con el dato
print(nuevo_df, "\n")

df = nuevo_df
# Eliminar una fila. Selecciona el índice deseado, en este caso el último
idx = df.index[df.shape[0]-1]

# El dataframe no se modifica. Crea una nueva instancia sin el dato
nuevo_df = df.drop(idx)
print(nuevo_df, "\n")

df = nuevo_df

# Cambia todas las filas que coinciden con "Peru" por "Perú"
# obtiene todas las coincidencias
coinciden = df["Pais"]=="Peru"

print(coinciden, "\n")  # las coincidencias tiene True en la fila coincidente

# modifica todas las filas
df[coinciden] = "Perú"
print(nuevo_df, "\n")
````
